```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, upper, when

# 1. Initialize the Spark Session
# appName identifies your job in the Spark UI
spark = SparkSession.builder \
    .appName("Som_Spark4_Lab") \
    .getOrCreate()

# 2. Create sample data (A list of tuples)
data = [
    ("postgresql", "relational", 1996),
    ("mongodb", "document", 2009),
    ("redis", "key-value", 2009),
    ("neo4j", "graph", 2007),
    ("cockroachdb", "distributed-sql", 2014)
]

# 3. Define column names and create the DataFrame
columns = ["DB_Name", "Type", "Release_Year"]
df = spark.createDataFrame(data, schema=columns)

# 4. Perform Transformations
# We'll make names uppercase and categorize them by age
df_transformed = df.withColumn("DB_Name", upper(col("DB_Name"))) \
                   .withColumn("Status", when(col("Release_Year") < 2010, "Classic").otherwise("Modern"))

# 5. Show Results (This triggers the actual computation)
print("--- Spark 4.1.0 Database Lab Results ---")
df_transformed.show()

# 6. Basic Aggregation: Count databases by status
df_transformed.groupBy("Status").count().show()

# Always stop the session when done
spark.stop()
```
